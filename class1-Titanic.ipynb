read_csv arguments:
    - filepath_or_buffer: the path of the file you want to read (receives a string, in case of windows a raw string -> "r''")
    - sep: the data separator (receives a string)
    - dtype: specifying the type of the data (you have to use np), if it is =None, the pandas will define for itself the type
    - header: used to specify wich line is going to be the head (receives an integer)
    - names: indicate the name of the columns (receives a list or an array), used when the head is no perfect
    - skiprows: skips an specified number of rows (receives an integer)
    - index_col: chooses a column to become the id of the dataset (receives an integer)
    - skip_blank_line: skipes blank lines (receives a boolean)
    - na_filter: eliminates the values of NaN in the dataset and leave it in blank (receives a boolean) -> very usefull

```python
data = pd.read_csv(filepath_or_buffer=r'C:\Curso de Machine Learning - udemy\datasets\titanic\titanic3.csv', sep=',', dtype=None,
         header=0, names=None, skiprows=1, index_col=2, skip_blank_lines=True, na_filter=False)
```
