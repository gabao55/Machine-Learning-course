# Reading data from an wikipedia URL

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

Acessing the html content from the webpage:

```python
medals_url = 'https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table'
response = requests.get(medals_url)
```

Parsing the html content:

```python
soup = BeautifulSoup(response.content, 'html5lib')
```

Searching and navigating through the parse tree:

```python
table = soup.find("table",{"class":"wikitable sortable"})
head = ['Country', 'Gold medals', 'Silver medals', 'Bronze medals', 'Total of medals']
df = {'Country': [], 'Gold medals': [], 'Silver medals': [], 'Bronze medals': [], 'Total of medals': []}
rows = table.find('tbody')

for row in rows.findAll('tr'):
    cells = row.findAll('td')
    if cells:
        df['Country'].append(cells[0].find('a').text)
        df['Gold medals'].append(cells[7].find(text=True).replace('\n', ''))
        df['Silver medals'].append(cells[8].find(text=True).replace('\n', ''))
        df['Bronze medals'].append(cells[9].find(text=True).replace('\n', ''))
        df['Total of medals'].append(cells[10].find(text=True).replace('\n', ''))
```

Converting the data extracted from wikipedia into a panda dataframe:

```python
df = pd.DataFrame(df)
df.head(20)
```

```python
df.to_csv(path_or_buf=r'..\datasets\exercise1\url_medals.csv')
```
