### Managing missing values in dataset:

Finding null and not null values:

```python
pd.isnull(data['body'])
```

```python
pd.notnull(data['body'])
```

Obtaining the amount of null values in a column:

```python
pd.isnull(data['body']).values.ravel().sum()
```

Missing values in a dataset could be caused by two reasons:
* Data extraction (incompatibility of data)
* Data recollection (data not available)

### Deleting missing values

There are two options:
* When there are just a few null values -> Delete the lines
* When there are too many null values -> Delete the column

The function used for deleting lines and columns with pandas is drop(), in which the parameter axis indicates if it is the line or the column (axis=0 -> delete the line ; axis=1 -> delete the column) and the parameter how indicates what values should be deleted (how='all' -> a column/line is deleted just if all elements contain Null values ; how='any' -> a column/line is deleted just if any element contains Null value).

```python
data.dropna(axis=0, how='all')
```

### Computing missing values 

Filling Null values:

```python
data2 = data
# It's not the best option, because there will be strings in numeric columns
data2.fillna('Unknown')
```

```python
# The best solution is switching the Null values by compatible values:
data3 = data
data3['body'] = data3['body'].fillna(0)
data3['home.dest'] = data3['home.dest'].fillna('Unknown')
data.tail()
```

```python
# Switching Null values by the column mean:
pd.isnull(data3['age']).values.ravel().sum()
```

```python
data3['age'] = data3['age'].fillna(data3['age'].mean().round())
data3.head(10)
```

```python
# Forward fill method (fills with the backward value):
data3['boat'] = data3['boat'].fillna(method='ffill')
data3.head(10)
```

```python
# Backward fill method (fills with the forward value):
data3['parch'] = data3['parch'].fillna(method='backfill')
data3.head(10)
```

Defining a function to create dummy variables in future:

```python
def create_dummies(df, var_name):
    dummy_var = pd.get_dummies(df[var_name], prefix=var_name)
    df = df.drop([var_name], axis=1)
    df = pd.concat([df, dummy_var], axis=1)
    return df
```
